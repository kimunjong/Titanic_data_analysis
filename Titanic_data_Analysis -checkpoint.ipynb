```python
# What is the total number of passengers in the dataset?
total_passengers = sum(df["Survived"].value_counts())
total_passengers
# total passenger = 891
```

```python
# How many passengers survived the disaster?
total_survived = df["Survived"].value_counts().plot(kind="pie") # 0 = don't survived, 1 = survived
total_survived
# total survived = 342
```

```python
df.groupby(["Sex", "Survived"])["Survived"].value_counts()
# total female servived = 233
# total male servived = 109
```

```python
sns.countplot(x="Survived", hue="Sex", data=df)
# show survival ratio of male and female
```
